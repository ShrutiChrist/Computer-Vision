## Non-Linear Spatial Filtering

### Loading the image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('/content/Dubai_Dubai-Marina-2017_(CC__BY-SA)Abubakr87210122-050135.jpg',0)
plt.imshow(img,"gray")
```

### Median Filter

A median filter is a non-linear digital filtering technique used for image processing. It is commonly employed to reduce noise in an image while preserving important edge details.

The median filter replaces each pixel's intensity value with the median intensity value from its local neighborhood.

```python
m, n = img.shape  #input dimesion
img_new1 = np.zeros([m, n]) # to store output of median filter, i had created an array of zeros
for i in range(1, m-1):
    for j in range(1, n-1):
#For each pixel at position (i, j), a 3x3 neighborhood is formed by extracting the intensity values of the pixel
#  and its surrounding pixels. These intensity values are stored in the temp list.
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= temp[4] #line assigns the median value of the sorted temp list
plt.imshow(img_new1,"gray")
```

### Max filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= temp[-1]  #max filter;  line assigns the maximum value in the temp lis
plt.imshow(img_new1,"gray")
```

### Min filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= temp[0]
plt.imshow(img_new1,"gray")
```

### Mid-point Filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= round((temp[0]+temp[-1])/2)
plt.imshow(img_new1,"gray")
```

### Alpha-trimmed Filter

```python
m, n = img.shape
img_new1 = np.zeros([m, n])
for i in range(1, m-1):
    for j in range(1, n-1):
        temp = [img[i-1, j-1],
               img[i-1, j],
               img[i-1, j + 1],
               img[i, j-1],
               img[i, j],
               img[i, j + 1],
               img[i + 1, j-1],
               img[i + 1, j],
               img[i + 1, j + 1]]
        temp = sorted(temp)
        img_new1[i, j]= round((sum(temp[1:-1]))/7)
plt.imshow(img_new1,"gray")
```
