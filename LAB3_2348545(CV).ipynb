the image negative is commonly defined as the inversion of pixel intensities in an image. This means that bright pixels become dark, and dark pixels become bright.

In OpenCV, you can obtain the negative of an image by subtracting each pixel intensity value from the maximum intensity value that the pixel type can represent. For example, for an 8-bit grayscale image, the maximum intensity value is 255. Thus, to obtain the negative of the image, you subtract the original intensity value from 255.

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Read the image
image = cv2.imread('/content/golden-retriever-royalty-free-image-506756303-1560962726.jpg', 0)  # Read the image as grayscale

# Perform image negative using NumPy
negative_image = 255 - image

# Display the original and negative images
cv2_imshow(image)
cv2_imshow(negative_image)
```

gray level slicing, also known as intensity level slicing, is a technique used to enhance specific intensity ranges within an image while suppressing others. The purpose of gray level slicing is to selectively highlight or manipulate pixels within a certain intensity range, often for visualization, analysis, or feature extraction purposes.

Example: X-Ray

```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Read the image
image = cv2.imread('/content/golden-retriever-royalty-free-image-506756303-1560962726.jpg', 0)  # Read the image as grayscale

# Define the intensity ranges for slicing (e.g., enhancing pixels with intensity values in the range [100, 200])
min_intensity = 100
max_intensity = 200

# Create a mask for pixels within the specified intensity range
mask = np.zeros(image.shape, dtype=np.uint8)
mask[(image >= min_intensity) & (image <= max_intensity)] = 255

# Apply the mask to the original image
sliced_image = cv2.bitwise_and(image, image, mask=mask)

# Display the original and sliced images using cv2_imshow
cv2_imshow(image)
cv2_imshow(sliced_image)
```

The technique works by effectively redistributing the pixel intensities in the image, such that the histogram of the resultant image is as uniform as possible.

This helps to maximize the dynamic range of the image, **enhancing the visibility of details and improving the overall image quality**.

```python
from google.colab.patches import cv2_imshow
import cv2

# Perform histogram equalization
equalized_image = cv2.equalizeHist(image)

# Display the original and equalized images using cv2_imshow
cv2_imshow(image)
cv2_imshow(equalized_image)
```
